```python
# Подключение библиотек

import os
import re
import numpy as np
import pandas as pd
import random as rd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
# Загрузка файлов

def load_texts(folder_path, author):
  texts = []
  for filename in os.listdir(folder_path):
    if author in filename.lower():
      with open(os.path.join(folder_path, filename),
                'r', encoding='utf-8') as file:
        texts.append(file.read())
  return texts

zoshchenko_texts = load_texts('/content/', 'zoshchenko')
gogol_texts = load_texts('/content/', 'karamzin')
```

```python
# Обработка и разбиение текстов на части

def clean_text(text):
  text = re.sub(r'[—–-](то|ка|что|за|нибудь)\b', '', text)
  text = re.sub(r'(\w+)-(\w+)', r'\1 \2', text)
  text = re.sub(r'\[.*?\]', '', text)
  text = re.sub(r'\b[IVXLCDM]+\.?\b', '', text, flags=re.IGNORECASE)
  text = re.sub(r'[^\w\s.!?]', '', text)
  text = re.sub(r'\.{3,}', '.', text)
  text = re.sub(r'\.{2,}', '', text)
  text = re.sub(r'[\r\n]+', ' ', text)
  text = re.sub(r'[ \t]+', ' ', text)
  text = text.replace('\xa0', '')
  text = text.strip()

  return text.lower()

def split_into_parts(text, target_length=1000):
  sentences = re.split(r'(?<=[.!?])\s+', text)
  parts = []
  current_part = []
  current_length = 0

  for sent in sentences:
    sent_words = sent.split()
    if current_length + len(sent_words) <= target_length:
      current_part.append(sent)
      current_length += len(sent_words)
    else:
      if current_part:
        parts.append(' '.join(current_part))
      current_part = [sent]
      current_length = len(sent_words)

  if current_part:
      parts.append(' '.join(current_part))

  return parts

zoshchenko_parts = []
for text in zoshchenko_texts:
  cleaned = clean_text(text)
  zoshchenko_parts.extend(split_into_parts(cleaned))

gogol_parts = []
for text in gogol_texts:
  cleaned = clean_text(text)
  gogol_parts.extend(split_into_parts(cleaned))

if zoshchenko_parts:
    zoshchenko_parts = zoshchenko_parts[:-1]

if gogol_parts:
    gogol_parts = gogol_parts[:-1]

zoshchenko_parts = rd.sample(zoshchenko_parts, 500) if zoshchenko_parts else []
gogol_parts = rd.sample(gogol_parts, 500) if gogol_parts else []
```

```python
len(gogol_parts)
```

```python
# Вычисление характеристик и создание датафрейма

def part_analysis(part):
  if not part.strip():
    return [0, 0]

  sentences = re.split(r'[.!?]+', part)
  sentences = [s.strip() for s in sentences if s.strip()]

  if not sentences:
    return [0, 0]

  total_s = len(sentences)
  total_w = 0
  total_c = 0

  for sentence in sentences:
    words = sentence.split()
    num_words = len(words)
    total_w += num_words
    total_c += sum(len(word) for word in words)

  avg_word_len = total_c / total_w if total_w else 0
  avg_sentence_len = total_w / total_s if total_s else 0

  return [avg_word_len, avg_sentence_len]

data = []

for part in zoshchenko_parts:
  features = part_analysis(part)
  data.append({'author': 'Zoshchenko',
               'avg_word_length': features[0],
               'avg_sent_length': features[1]})

for part in gogol_parts:
  features = part_analysis(part)
  data.append({'author': 'Karamzin',
               'avg_word_length': features[0],
               'avg_sent_length': features[1]})

df = pd.DataFrame(data)
```

```python
# Построение графика

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df,
                x='avg_word_length',
                y='avg_sent_length',
                hue='author',
                alpha=0.7)
plt.title('Scatterplot of Text Features by Author')
plt.xlabel('Average Word Length')
plt.ylabel('Average Words per Sentence')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Разделение данных и классификация

df['author_label'] = df['author'].map({'Zoshchenko': 0, 'Karamzin': 1})

X = df[['avg_word_length', 'avg_sent_length']]
y = df['author_label']
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3,
                                                    random_state=42)
model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
